```
# Lib
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# ***(a) To begin, load in the Boston data set. The Boston data set is part of the MASS library in R. How many rows are in this data set? How many columns? What do the rows and columns represent?***

```
Boston_data = pd.read_csv("Boston.csv")
print(Boston_data)
print("- The number of data set rows: ",Boston_data.shape[0])
print("- The number of data set columns: ",Boston_data.shape[1])  
print("- What the dataset shows the values of : ", list(Boston_data.columns.values))
```

```
sns.pairplot(Boston_data)
```

# ***(d) Do any of the suburbs of Boston appear to have particularly high crime rates? Tax rates? Pupil-teacher ratios? Comment on the range of each predictor.***

```
fig, axes = plt.subplots(nrows = 1, ncols = 3, figsize = (20, 5))
sns.distplot(Boston_data["CRIM"], kde = False, ax = axes[0])
axes[0].set(xlabel = "", title = "Histogram of per capita crime rate")

sns.distplot(Boston_data["TAX"], bins = 20, kde = False, ax = axes[1])
axes[1].set(xlabel = "", title = "Histogram of Tax rate")

sns.distplot(Boston_data["PTRATIO"], kde = False, ax = axes[2])
axes[2].set(xlabel = "", title = "Histogram of Pupil-teacher ratio")

Boston_data.loc[:, ["CRIM", "TAX", "PTRATIO"]].describe()
```

```
print("The number of the suburbs average more than seven rooms per dwelling is: ", len(Boston_data[Boston_data["RM"]>7]["RM"]))
print("The number of the suburbs average more than eight rooms per dwelling is: ", len(Boston_data[Boston_data["RM"]>8]["RM"]))
print(Boston_data.loc[Boston_data["RM"] > 8])
print(Boston_data.loc[Boston_data["RM"] > 8].describe().round(4))
```
